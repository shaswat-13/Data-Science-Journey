## Decorators

Decorators are basically wrapper functions that extend the functionality/ behaviour of other  base functions.

### Why decorators?
- a single function should have only one task so decorators extend the behaviour
- code modularity
- code reusability

```python
import time
# helps to maintain the metadata of original function
from functools import wraps 
```

```python
# a decorator to calculate the time taken to execute a function
def time_execution(func):
    @wraps(func)
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        total_time = end_time - start_time
        print(f"It took {round(total_time,5)}s.")
    return wrapper
```

```python
# a decorator to only execute the function if the user has admin privileges
def requires_admin(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if is_admin:
            func(*args, **kwargs)
        else:
            print("Admin Privileges Required!")
    return wrapper
```

```python
# keeps track of how many times wrapper function has been called
def count_calls(func):
    @wraps(func)
    def wrapper():
        wrapper.count += 1
        func()
        print(f"Total function calls: {wrapper.count}")
    return wrapper
```

### When no arguements

```python
def brew_tea():
    print('Tea Brewing...')
    time.sleep(2)
    print('Tea Ready!!!')

# normal call
brew_tea()
```

```python
# decorator call with function as an input
# used when based on dynamic state, either extended (decorated) function should be called
# or simply normal function
timed_tea_brew = time_execution(brew_tea)

condition = True
if condition:
    timed_tea_brew()
else:
    brew_tea()
```

```python
# decorator with @ 
# used for explicitly enhancing the behaviour of a function
@time_execution
def make_matcha():
    print('Making matcha...')
    time.sleep(3)
    print('Matcha Ready!!!')

make_matcha()
```

### When positional or keyword arguements


```python
@requires_admin
def make_decisions(name, is_admin=False):
    if is_admin:
        print(f'{name} just made a decision.')

make_decisions(name='Ram', is_admin=False)
```
