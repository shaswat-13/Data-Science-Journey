```python
import random

def generate_messy_txt(filename, word_count):
    # A mix of common words to ensure your frequency counter actually finds "trends"
    words = ["apple", "banana", "data", "python", "code", "logic", "frequency", 
             "test", "random", "string", "hello", "world", "AI", "loop"]
    
    # Generate random words + some filler
    content = []
    for _ in range(word_count):
        word = random.choice(words)
        # Randomly add punctuation to test your cleaning logic!
        if random.random() > 0.8:
            word += random.choice([".", ",", "!", "?"])
        content.append(word)
    
    with open(filename, 'w') as f:
        f.write(" ".join(content))
    print(f"Created {filename} with {word_count} words.")

# Generate your three files
generate_messy_txt("sample_500.txt", 500)
generate_messy_txt("sample_1000.txt", 1000)
generate_messy_txt("sample_1500.txt", 1500)
```
