```python
filename = 'sample_1500.txt'
content = str()
with open(filename, 'r') as f:
    content = f.read()
print(content)
```

```python
# replace the punctuations using maketrans method of str
punctuations = ".,?!"
table = str.maketrans('', '', punctuations)
cleaned_content = content.translate(table)
cleaned_content = cleaned_content.lower()
```

```python
list_of_words = cleaned_content.split()
word_count = {}
total_word_count = 0
for word in list_of_words:
    if word not in word_count:
        word_count[word] = 0

    word_count[word] += 1
    total_word_count += 1
print(word_count)
print(total_word_count)
    
```

```python
##### BETTER WAY

filename = 'sample_1500.txt'

with open(filename, 'r') as f:
    content = f.read()

# 1. Cleaning
punctuations = ".,?!"
table = str.maketrans('', '', punctuations)
cleaned_content = content.translate(table).lower()

# 2. Splitting (using .split() catches newlines and extra spaces)
list_of_words = cleaned_content.split()

# 3. Counting
word_count = {}
for word in list_of_words:
    word_count[word] = word_count.get(word, 0) + 1

# 4. Displaying (Sorting by frequency)
# This sorts the dictionary by values in descending order
sorted_words = sorted(word_count.items(), key=lambda item: item[1], reverse=True)

print(f"Total Word Count: {len(list_of_words)}")
print("Top 10 Words:")
for word, count in sorted_words[:10]:
    print(f"{word}: {count}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

most_frequent_words = sorted_words[:10]
df = pd.DataFrame(most_frequent_words)
df.columns = ['word', 'count']
df.head()
```

```python
sns.barplot(data=df, x="word", y="count")
plt.title("Ten Most Repeated Words and their WordCounts")
plt.show()
```
