### if a number is divisible by 3 only, it displays Fizz, if its divisible by 5 only, it displays Buzz and if its divisible by both 3 and 5, it displays FizzBuzz

```python
import random
import numpy as np

number = random.randint(1,315)

print(number)
```

```python
op = ""

if number % 3 == 0:
    op += "Fizz"
if number % 5 == 0:
    op += "Buzz"

print(op)
```
