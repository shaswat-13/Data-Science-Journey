```python
class Calculator():
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    
    def add(self):
        return self.num1+self.num2
    
    def subtract(self, rev=False):
        if rev == True:
            return self.num2 - self.num1
        return self.num1 - self.num2
    
    def multiply(self):
        return self.num1*self.num2
    
    def divide(self, rev=False):
        try:
            if rev == True:
                q = self.num2/self.num1
            else:
                q = self.num1/self.num2
        except ZeroDivisionError:
            print('Cant divide by 0')
            return None
        else:
            return round(q,2)
    
    def display_results(self):
        print(f"Add: {self.add()}")
        print(f"Subtract: {self.subtract()}")
        print(f"Subtract: {self.subtract(rev=True)}")
        print(f"Multiplication: {self.multiply()}")
        print(f"Division: {self.divide()}")
        print(f"Division: {self.divide(rev=True)}")

        
```

```python
c = Calculator(5,0)
c.display_results()
```
