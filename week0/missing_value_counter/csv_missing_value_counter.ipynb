```python
import pandas as pd
import numpy as np

df = pd.read_csv('messy_data.csv')
df.head()
```

```python
# Convert empty strings or just spaces to True NaN objects
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
```

```python
total_rows = len(df['ID'])
for column in df.columns:
    complete_rows = df[column].count()
    missing_rows = total_rows - complete_rows
    print(f"There are {missing_rows} missing values in {column} column.")
```

```python
# using pandas method
missing_counts = df.isnull().sum()
print(missing_counts)
```
