```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

rows = 100
data = {
    'ID': range(1, rows + 1),
    'Date': pd.date_range('2026-01-01', periods=rows).strftime('%Y-%m-%d'),
    'Temperature': np.random.uniform(10, 35, size=rows).round(1),
    'Humidity': np.random.randint(30, 80, size=rows),
    'City': np.random.choice(['London', 'New York', 'Paris', 'Tokyo'], size=rows),
    'Status': np.random.choice(['Active', 'Inactive'], size=rows),
    'Sensor_Code': [f'SN{i:03}' for i in range(1, rows + 1)]
}

df = pd.DataFrame(data)

# Randomly replace ~15% of the data with NaN (Nulls)
for col in df.columns:
    if col != 'ID': # Keep ID complete
        df.loc[df.sample(frac=0.15).index, col] = np.nan

df.to_csv('messy_data.csv', index=False)
print("messy_data.csv created with missing values!")
```
