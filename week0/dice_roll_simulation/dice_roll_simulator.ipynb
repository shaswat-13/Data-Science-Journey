```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
random.seed(67)
N = 1000
tracker = {}

for i in range(N):
    dice_roll_1 = random.randint(1,6)
    dice_roll_2 = random.randint(1,6)
    sum = dice_roll_1+dice_roll_2
    if sum not in tracker:
        tracker[sum] = 0
    tracker[sum] += 1

print(tracker)
```

```python
list_of_tracker = list(tracker.items())
df = pd.DataFrame(list_of_tracker)
df.columns = ['dice roll sum', 'frequency']
df.head()
```

```python
sns.barplot(data=df, x="dice roll sum", y="frequency")
plt.title(f"Distribution of Dice Roll Sums for {N} times")
plt.show()
```
