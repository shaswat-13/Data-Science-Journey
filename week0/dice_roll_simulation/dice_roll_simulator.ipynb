```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
random.seed(67)
N = 1000
tracker = {}

for i in range(N):
    dice_roll = random.randint(1,6)

    if dice_roll not in tracker:
        tracker[dice_roll] = 0
    tracker[dice_roll] += 1

print(tracker)
```

```python
list_of_tracker = list(tracker.items())
df = pd.DataFrame(list_of_tracker)
df.columns = ['dice roll', 'frequency']
df.head()
```

```python
sns.barplot(data=df, x="dice roll", y="frequency")
plt.title(f"Distribution of Dice Roll for {N} times")
plt.show()
```
