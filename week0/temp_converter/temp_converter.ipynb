```python
import pandas as pd
import numpy as np

df = pd.read_csv('temp_C.csv')
df.head()
```

```python
def convert_c2fk(df):
    new_df = pd.DataFrame()
    new_df['Date'] = df['Date']
    new_df['Celcius'] = df['Temperature']
    new_df['Fahrenheit'] = round((df['Temperature']*1.8+ 32),2) 
    new_df['Kelvin'] = df['Temperature'] + 273
    return new_df

def convert_f2ck(df):
    new_df = pd.DataFrame()
    new_df['Date'] = df['Date']
    new_df['Fahrenheit'] = df['Temperature']
    new_df['Celcius'] = round((df['Temperature'] - 32)/1.8,2)
    new_df['Kelvin'] = new_df['Celcius'] + 273
    return new_df

def convert_k2fc(df):
    new_df = pd.DataFrame()
    new_df['Date'] = df['Date']
    new_df['Kelvin'] = df['Temperature']
    new_df['Celcius'] = df['Temperature'] - 273
    new_df['Fahrenheit'] = round((new_df['Celcius']*1.8+32),2) 
    return new_df
```

```python

def temp_converter(df, degree):
    '''Input:
        - df: dataframe with columns Date and Temperature
        - degree: c,k,f signifying Celcius, Kelvin or Fahrenheit
        
        Output:
        - df: data frame with one degree converted to other two'''
    degree = degree.lower()
    if degree == "k":
        new_df = convert_k2fc(df)
    elif degree == "c":
        new_df = convert_c2fk(df)
    elif degree == "f":
        new_df = convert_f2ck(df)
    return new_df

converted_df = temp_converter(df, 'C')
converted_df.to_csv('converted.csv')
```
